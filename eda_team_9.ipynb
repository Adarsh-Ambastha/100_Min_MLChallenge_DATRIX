```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
dt['Risk'] = dt['Risk'].str.strip()
```

```python
numerical_columns = ['1 month return', 'NAV', '1 Year return', '3 Year Return', 'AUM']

corr_matrix = dt[numerical_columns].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Mutual Fund Variables', fontsize=15)
plt.show()
```
